```python
from random import*
def x(n):
    L=[]
    d=0
    for i in range(n):
        r=random()
        if r<=0.1:
            d=10
        elif r<=0.3:
            d=20
        elif r<=0.6:
            d=30
        else:
            d=40
        L.append(d)
    return L
            
```

```python
L=x(100000)
print(L)
```

```python
print(sum(L)/len(L))
```

```python
from random import*
def x(n):
    M=[]
    d=0
    for i in range(n):
        r=random()
        if r<=1/2:
            d=0
        elif r<=5/6:
            d=2
        else:
            d=1
        M.append(d)
    return M
            
```

```python
M=x(1000)
print(M)
```
