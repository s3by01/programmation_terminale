```python
from math import*
def Decomp(n):
    L=[]
    p=2
    while n%p==0:
        L.append(2)
        q=int(n/p)
        n=q
    d=3
    while d<=n:
        while n%d==0:
            L.append(d)
            q=int(n/d)
            n=q
        d=d+2
    return L
```

```python
Decomp(486625396124862)
```

```python
for k in range(100):
    nb=(k+3)**2
    print(k,nb,Decomp(nb))
```
